```python
%%time
# Separate majority and minority classes
df_majority = test[test.default_payment_next_month==0]
df_minority = test[test.default_payment_next_month==1]
 
```

```python
%%time
# Upsample minority class
#df_minority_upsampled = resample(df_minority, replace=True, n_samples=351976183,random_state=123) 
```

```python
%%time
df_minority_upsample = df_minority.sample(len(df_majority),replace='True')
```

```python
# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_majority, df_minority_upsample])
 
# Display new class counts
df_upsampled.default_payment_next_month.value_counts()
#1    23364
#0    23364
#Name: default_payment_next_month, dtype: int64
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
```

```python
drop_col=['ID','default_payment_next_month']
X = df_upsampled.drop(drop_col,axis=1)
y = df_upsampled['default_payment_next_month']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=1)
```

```python
model=RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```
